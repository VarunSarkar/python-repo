<h2>Matplotlib - 2</h2>

<p><h3>Q. You work in XYZ Corporation as a Data Analyst. Your corporation has told you to visualize the 
mtcars.csv dataset with various plots.</h3>
<h3>1. Generate a bar plot graph for the columns ‘carbs’ and ‘gear’. </h3>
</p>
<p><h4>a. Map the ‘carbs’ onto the x-axis.  <br>
b. Map the ‘gear’ onto the y-axis. <br>
c. Provide the x-axis label as Number of carburetors.  <br>

```python
carbs = cars['carb']
gears = cars['gear']
```

```python
fig = plt.figure(figsize=(10,6))

plt.bar(carbs, gears)
plt.xlabel("Number of carburetors")
plt.ylabel("Number of forward gears")
plt.title("carbs vs gear")
plt.show()
```

<h2>Matplotlib - 3</h2>

</p>
<p><h4>a. Map the ‘wt’ onto the x-axis. <br>

c. Provide the y-axis label as ‘Count’  <br>
